# ABS/MBS Analysis

The following code is intended to provide some insights on the ABS universe, and the impact of including it as part of the Fixed Income Portfolio. First, I load the libraries we will need to run the analysis:

```python
import analytics_hero.fundAnalytics as fundAnalytics
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
ac = fundAnalytics.AnalyticsClass()
plt.style.use('seaborn-deep')
sns.set_color_codes()
```

The first thing I will do, is to take a look at the behavior of the cumulative returns of different Fixed Income indices:

```python
ret = ac.get_performance([12571, 1221120, 12211319, 31219], init='2012-01-01', end='2017-07-01')
names = ac.get_table('identifiers')
fund_name = []
for col in ret.columns:
    fund_name.append(names['Product_Name'][names['Product_ID'] == col].values[0])
ret.columns = fund_name
ret.cumsum().plot()
plt.show()
```
